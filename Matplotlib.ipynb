```python
## Create multiple plots with subplots
```

```python
names = ['Pratibha','HRushikesh', 'Varsha', 'Ayan', 'Vilas', 'Sayali']
scores = [90,95,97,91,93,95]
```

```python
plt.figure(figsize=(20,10))

plt.subplot(133)
plt.bar(names, scores)

plt.subplot(132)
plt.scatter(names , scores)

plt.subplot(131)
plt.plot(names , scores)

plt.suptitle("Scores of students in Exam")

plt.show()
```

```python
names = ['Pratibha','HRushikesh', 'Varsha', 'Ayan', 'Vilas', 'Sayali']
scores = [90,95,97,91,93,95]

plt.figure(figsize=(20,10))

plt.subplot(311)
plt.bar(names, scores)

plt.subplot(312)
plt.scatter(names , scores)

plt.subplot(313)
plt.plot(names , scores)

plt.suptitle("Scores of students in Exam")

plt.show()
```
